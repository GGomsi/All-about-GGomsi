```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# loading data

df= pd.read_csv('kidney_disease.csv')
df.head()
```

```python
# 26개의 col수와 400개의 index가 있구나
df.shape
```

```python
# id col은 예측하는데 필요없어 그래서 삭제할거야
df.drop('id', axis = 1, inplace = True)
```

```python
# 친숙한 col명으로 바꿀거야
df.columns = ['age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell',
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
              'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
              'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema',
              'aanemia', 'class']
```

```python
# 어? 근데 이건 숫자잖아 따라서 object 지만 숫자로 형 변환을 해보자
df['packed_cell_volume']
```

```python
df['white_blood_cell_count']
```

```python
# 이건 심지어 NaN 값도 있네
df['red_blood_cell_count']
```

```python
# object 타입을 숫자로 변형시켜줄거야 to_numeric을 쓸거고
# 스칼라값, list, tuple, Series 등이 대상으로 지정될 수 있어
# errors = 'ignore' -> 만약 숫자로 변경할 수 없는 데이터라면 숫자로 변경하지 않고 원본 데이터를 그대로 반환합니다.
# errors = 'coerce' -> 만약 숫자로 변경할 수 없는 데이터라면 기존 데이터를 지우고 NaN으로 설정하여 반환합니다.
# errors = 'raise' -> 만약 숫자로 변경할 수 없는 데이터라면 에러를 일으키며 코드를 중단합니다.

df['packed_cell_volume'] = pd.to_numeric(df['packed_cell_volume'], errors='coerce')
df['white_blood_cell_count'] = pd.to_numeric(df['white_blood_cell_count'], errors='coerce')
df['red_blood_cell_count'] = pd.to_numeric(df['red_blood_cell_count'], errors='coerce')
```

```python
# 이제 숫자였지만 object로 표현되었던 데이터들은 전부 float로 바꾸었어
df.info()
```

```python
df.columns
```

```python
# 이제 나머지 object들을 꺼내야돼(Extract)
# object 타입은 cat_cols에 넣고
cat_cols = [col for col in df.columns if df[col].dtype == 'object']
# 나머지 float 타입은 num_cols에 넣어보자
num_cols = [col for col in df.columns if df[col].dtype != 'object']
```

```python
# object 타입의 col들 어떻게 구성되어 있는지 쭉 볼거야
# 흠... 지저분한 데이터들이 있네? 이런건 전부 깔끔하게 바꿔야돼
for col in cat_cols:
    print(f"{col} has {df[col].unique()} values\n")
```

```python
# to_replace를 사용하면 dict(Map)처럼 값을 변경하기 쉽구나
df['diabetes_mellitus'].replace(to_replace = {'\tno':'no','\tyes':'yes',' yes':'yes'},inplace=True)

df['coronary_artery_disease'] = df['coronary_artery_disease'].replace(to_replace = '\tno', value='no')

df['class'] = df['class'].replace(to_replace = {'ckd\t': 'ckd', 'notckd': 'not ckd'})
```

```python
# 자 이제 class은 label 값이기 때문에 0과 1로 바꾸어 줘야 돼
df['class']
```

```python
# 근데 여기서 보면 map을 이용해서 0과 1로 바꾸었지만 object 타입이란말이지
# 그래서 확실하게 float으로 지정해주기 위해서 to_numeric을 써
df['class'] = df['class'].map({'ckd': 0, 'not ckd': 1})
df['class'] = pd.to_numeric(df['class'], errors='coerce')
```

```python
# 이제 바뀐 class와 나머지 object들을 볼까?
cols = ['diabetes_mellitus', 'coronary_artery_disease', 'class']

for col in cols:
    print(f"{col} has {df[col].unique()} values\n")
```

```python
# 사실상 우리는 여기서 class 부분만 label에 추가할거야 
plt.figure(figsize = (20, 15))
plotnumber = 1

for column in num_cols:
    if plotnumber <= 14:
        ax = plt.subplot(3, 5, plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column)
        
    plotnumber += 1

plt.tight_layout()
plt.show()
```
