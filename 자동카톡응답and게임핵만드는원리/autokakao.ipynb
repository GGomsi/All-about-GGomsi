```python
import pyautogui
import pyperclip
import time
import threading
import os

os.chdir(os.path.dirname(os.path.abspath('autokakao.ipynb')))

def send():
    threading.Timer(10, send).start()
    
    picPosition = pyautogui.locateOnScreen('1.png')
    print(picPosition)
    
    if picPosition is None:
        picPosition = pyautogui.locateOnScreen('2.png')
        print(picPosition)
    
    if picPosition is None:
        picPosition = pyautogui.locateOnScreen('3.png')
        print(picPosition)
    
    clickPosition = pyautogui.center(picPosition)
    pyautogui.doubleClick(clickPosition)
    
    pyperclip.copy("세번째 테스트 내이름 부르면 해위~")
    pyautogui.hotkey("ctrl", "v")
    time.sleep(1.0)
    
    pyautogui.write(["enter"])
    time.sleep(1.0)
    
    pyautogui.write(["escape"])
    time.sleep(1.0)

send()
```

```python
# fps 핵 만들기
# https://simritest.com/reaction/
import pyautogui
import pyperclip
import time
import os

# 경로를 .py파일의 실행경로로 이동, 현재 경로로 이동
os.chdir(os.path.dirname(os.path.abspath('autokakao.ipynb')))

n = 0
while(n<5):
    # 모니터 기준 사진 좌표, 가로길이 세로길이를 찍어준다.
    # 픽셀수가 적고, 단색의 이미지일수록 더 빠르다.
    # 모니터 기준 사진 좌표(left,top), 가로길이(width) 세로길이(height를 찍어준다.
    picPosition = pyautogui.locateOnScreen('click1.png')
    print(picPosition)
    if picPosition != None:
        # 마우스를 이동시켜준다(picPosition.left는 좌우, top은 상하)
        # 사진 기준으로 0,0으로 이동
        # 사진이 없으면 좌표를 못찾아서 오류가 난다.
        pyautogui.moveTo(picPosition.left, picPosition.top)
        
        # 클릭한다. default 값은 좌클릭
        pyautogui.click()
        n += 1
```
